# Zurückgelegter Weg in Abhängigkeit zur der Dauer des Zeitfensters

## Einleitung

Das Verhältnis zwischen zurückgelegter Strecke in $m$ während Messwerte mit den Smartphone Sensoren erfasst werden und Geschwindigkeit in $km/h$ wird in einem Diagramm dargestellt. Zu zeigen ist, dass die Länge $l$ der gefahrenen Strecke abhängig ist von der Dauer des Zeitfensters in $ms$. Die Annahme ist, dass bei der Wahl eines dynamischen Zeitfensters die zurückgelegte Strecke $l$ auf einen Meter normiert werden kann siehe Gleichung 2. Zudem wird ein statisches Zeitfenster der Dauer $t=500ms$ und $t=80ms$ (siehe G. 1) untersucht. 

$$
l = v * t \Rightarrow m = \frac{m}{s}  *  \frac{Zeitfenster\; t\; in\; ms}{1000}
$$ <div style="text-align: right"><i>(1)</i></div>

Beispiel Gleichung 1: Geg.: Geschwindigkeit $v = 20km/h$, Dauer des statischen Zeitfensters $ t = 500ms$   

$$
m = \frac{18\frac{km}{h}}{3.6}  *  \frac{500ms}{1000}
$$   
<br>    
<div style="text-align: center">$m = 5\frac{m}{s} * 0.5s = $ \underline{2,5m}</div>
<br>
<br>

$$
l = v * \frac{l}{v} \Rightarrow m = \frac{m}{s} * \frac{1m}{\frac{m}{s}} 
$$ <div style="text-align: right"><i>(2)</i></div>

Beispiel Gleichung 2: Geg.: Geschwindigkeit $ v = 20km/h$

$$
m = \frac{18\frac{km}{h}}{3.6}  *  \frac{1m}{\frac{18\frac{km}{h}}{3.6}} 
$$ 
<br>
<div style="text-align: center">$m = 5\frac{m}{s}  *  0.2s =  $ \underline{1m}</div>
<br>
<br>


Bei der Berechnung des dynamischen Zeitfensters wird berechnet wie viel $ms$ nötig sind, um $1m$ Weg zurück zu legen (siehe G. 3). 

$$
t = \frac{l}{v} \Rightarrow ms = \frac{1m}{\frac{m}{s}} * 1000 
$$ <div style="text-align: right"><i>(3)</i></div>



## Darstellung des zurückgelegten Wegs in Abhängigkeit der Geschwindigkeit und der Dauer des Zeitfensters  

Auf Abb. 1 ist zu sehen, wie sich die Wahl der Dauer des Zeitfensters $t$ und die Geschwindigkeit $v$ auf die Länge $l$ des zurückgelegten Wegs auswirkt. Die X-Achse zeigt die Geschwindigkeit $v$ in $km/h$ und die Y-Achse den zurückgelegten Weg $l$ in $m$. Zu sehen sind drei lineare Funktionen, die abhängig sind von der Wahl der Zeit $t$ des Zeitfensters. Die Idee hinter den statischen Zeitfenstern ist der triviale Ansatz, dass keine Berechnung nötig ist, welche abhängig von der Geschwindigkeit ist, um das die Dauer $t$ des zeitfensters zu bestimmen. Ein Zeitfenster mit einer Dauer von $t=500ms$ bietet den Vorteil mehr Messwerte zu erfassen zu einer längeren Wegstrecke und damit Muster besser zu erkennen z.B. Pflasterfugen bei Gehwegplatten. Mit einem statischen Zeitfenster $t$ von $500ms$ steigt der zurückgelegte Weg linear an. Die Steigung $m$ beträgt $m=\frac{
\Delta y}{\Delta x} = \frac{y_1-y_0}{x_1-x_0} \Rightarrow \frac{5-2,5}{36-18} = 0,938$.  Dagegen hat ein statisches Zeitfenster von 80ms eine kleinere Steigung von $m=\frac{
\Delta y}{\Delta x} \Rightarrow \frac{0,8-0,4}{36-18} = 0,022$. Dies bedeutet das mit einem Zeitfenster von $80ms$ die zurückgelegte Strecke relativ klein ist.

Der verwendete Bildausschnitt zeigt ca. $1m$ ($\approx97cm$) Weglänge siehe Abb. 2. Die untere Begrenzung auf $\approx47cm$ ist nötig, weil auf einigen Bildern der Schuh des Radfahrers zu sehen ist. Hiermit wird versucht diese für das Faltungsnetz nicht wichtige Information zu vermeiden. Mit einem Zeitfenster von $80ms$ werden weniger Sensormesswerte erfasst als Wegfläche auf den Bildausschnitt zu erkennen ist bei Geschwindigkeiten kleiner als $v=45km/h$. Der Vorteil bei einem kleinen Zeitfenster ist die Bildrate von ca. $3,4$ Bildern pro Sekunde. Zum Vergleich bei einem Zeitfenster mit nicht optimierte Berechnung der Wartezeit beträgt die Bildrate ca. $\approx0,9$ Bilder pro Sekunde und $\approx2,248$ Bilder pro Sekunde mit den Optimierungen siehe das Jupyter Notebook "frameRate". Hieraus entstand die Motivation ein dynamisches Zeitfenster zu verwenden.

Abbildung 2: Bildausschnitt mit der Nutzung des dynamischen Zeitfensters

Die Dauer des dynamischen Zeitfensters berechnet sich aus der Gleichung 3. Das heißt die Dauer $t$ ist Abhängigkeit von der aktuellen Geschwindigkeit v des Fahrrads. Wie auf Abb. 1 zu sehen entspricht der zurückgelegte Weg dann immer $1m$. Dafür muss die Voraussetzung erfüllt sein, dass die aktuelle Geschwindigkeit möglichst genau ist. Um die aktuelle Geschwindigkeit zu erhalten wird die Android Location API genutzt sowie der Location Manager. Der Location Manager aktualisiert die Aktuelle GPS Lokation so schnell wie möglich. Dabei wird in der API nicht beschrieben mit welcher Frequenz. Ein weiteres Problem ist, dass nicht angegeben wird wie die Geschwindigkeit gemessen wird. In dem Buch *Essentials of positioning and location technology* wird beschrieben das meistens der Doppler-Effekt genutzt wird für die Berechnung der Geschwindigkeit.  

Als negatives Beispiel wird angenommen, dass die aktuelle Geschwindigkeit $v = 20km/h$ ist aber die gemessene $16km/h$ beträgt. Dann ist das Zeitfenster der gemessenen Geschwindigkeit $t=228ms$ wobei das eigentliche Zeitfenster $180ms$ dauern müsste. In diesem theoretischen Beispiel werden also Sensormesswerte erfasst für den zurückgelegten Weg von $m = \frac{20}{3,6}*0.228s = 1.2667$. Also ist es möglich, dass irrelevante Informationen über den Bildausschnitt erfasst werden können.  Die Android Location API ermöglicht es auch die Genauigkeit der aktuellen Geschwindigkeit anzugeben. Dazu wird die Methode getSpeedAccuracyMetersPerSecond() genutzt. Diese gibt einen float Wert zurück. Mit $68%$ Wahrscheinlichkeit befindet sich dann die berechnete Geschwindigkeit in diesem Bereich mit der Abweichung des float Wert. Hat z.B. der Aufruf der Methode einen Rückgabe Wert von $2$ ergeben und die berechnete Geschwindigkeit ist $20km/h$, dann liegt die Wahre Geschwindigkeit mit $68%$ Wahrscheinlichkeit in dem Bereich $18-22km/h$. Bisher wurde die Methode noch nicht genutzt.
Einen weiteren Rückschluss über die Aussagekraft bietet der Zeitstempel der GPS Lokation. Dieser wurde bisher auch noch nicht untersucht.

Die Dauer des dynamischen Zeitfensters ist beschränkt auf das abgeschlossenes Intervall $[ 9,432ms ,720ms ] := \lbrace x ∈ R ∣ 9ms ≤ x ≤ 720ms \rbrace$. Die obere Einschränkung ergibt sich daraus das die minimale gemessene Geschwindigkeit für die Aufnahme von Bildern $5km/h$ betragen muss dann ergibt sich ein Zeitfenster $t=720ms$. Während das Fahrrad steht sollen nämlich keine Bilder aufgenommen werden, um die Klassifizierung mit einem Faltungsnetz zu erleichtern. Die untere Grenze ist nur eine theoretische Grenze.  Bei einer Geschwindigkeit von $v = \frac{l}{t} \Rightarrow km/h = \frac{1m}{9ms}*3.6 \approx 382 km/h$ wäre dieser Schwellwert erreicht. Eine Geschwindigkeit die mit einem Fahrrad wahrscheinlich nicht erreicht werden kann. Für das verwendete Smartphone Nokia 6 beträgt die minimale mögliche Belichtungszeit $9432ns$. Die Belichtungszeit Hardware Grenzwerte können mit CameraCharacteristics.SENSOR_INFO_EXPOSURE_TIME_RANGE abgefragt werden. Diese Belichtungszeit ist auch der Schwellwert für die minimale Dauer des Zeitfensters. Hier bietet es sich auch an die untere Grenze auf die Top Geschwindigkeiten der Tour de France zu setzen. 

Als Alternative zu der Nutzung der Location API kann ein externer Geschwindigkeitssensor genutzt werden. Denkbar wäre ein Gerät mit einem Bluetooth Schnittstelle. Ein komplett anderer Ansatz wäre ein Embedded System zu entwickeln. Dieses System könnte mit einer Kamera verbunden sein, die am Unterrohr des Fahrrads befestigt ist. Auch Echtzeitanforderungen können besser erfüllt werden als mit einem Android Smartphone. Die Entwicklung ist allerdings komplexer und ein Smartphone ist ein kompaktes System mit vielen Sensoren, welches relativ einfach an einem Fahrrad montiert werden kann. Darüber hinaus besitzen viele Menschen ein Smartphone.

## Dynamisches Zeitfenster und das Weg-Zeit-Gesetz 

Bei der Berechnung des dynamischen Zeitfensters (G. 3) wird von einer gleichförmigen Bewegung ausgegangen.  
Damit ist eine gleichbleibende Geschwindigkeit gemeint. Dies trifft beim Fahrradfahren zum Teil während der Fahrt zu. Wenn der Fahrradfahrer das Fahrrad beschleunigt oder abbremst, dann ist dies keine gleichförmige Bewegung. Auch Unebenheiten können die Geschwindigkeit reduzieren. Es ändert sich also die positive oder negative Beschleunigung $a$. Bei den erfassten Datensätzen ist die maximale Gescheindigkeit niedriger als $45km/h$  

$s = \frac{1}{2} · a · t^2 + v_0 · t + s_0$
<div style="text-align: right"><i>(4)</i></div>

Hier wird die Überlegung betrachtet über das Weg-Zeit-Gesetz der gleichmäßig beschleunigten Bewegung siehe G. 4 das dynamische Zeitfenster zu berechnen. Dabei steht $s$ für die zurückgelegte Strecke, $a$ für die Anfangsbeschleunigung, $v$ die Anfangsbeschleunigung, $t$ Zeit und $s_0$ für den Anfangsweg. Die Gleichung 4 muss umgestellt werden auf $t$.  

Diese Herangehensweise hat den Nachteil, dass das Fahrrad auch nicht gleichmäßig beschleunigt. Dazu kommt auch noch die Problematik die Anfangsgeschwindigkeit und Anfangsbeschleunigung zu messen. Die Anfangsbeschleunigung $v_0$ kann mit der Y-Achse des Beschleunigungssensors des Smartphones gemessen werden. Zu beachten ist, wenn das Fahrrad und somit das Smartphone geneigt ich also der Nick Winkel nicht $0^\circ$ beträgt dann nehmen auch andere Beschleunigungsrichtungen einfluss auf die Y-Achsen Beschleunigung $v$.


## Fazit

Ein dynamisches Zeitfenster hat gegenüber einem statischen Zeitfenster den Vorteil zu einer gegebenen Geschwindigkeit $v$ des Fahrrads die ungefähre gleiche Strecke zurück zu legen (siehe Abb. 1). In diesem Fall der Datenerfassung sollen Sensormesswerte für den zurückgelegten Weg von $1m$ erfasst werden. Dies entspricht mit einer Abweichung von 3cm auch den erfassten Bildausschnitt. Das dynamische Zeitfenster hat den Nachteil, dass die Berechnung abhängig ist von der Genauigkeit der gemessenen Geschwindigkeit. Als alternative kann externer Sensor verwendet werden, der z.B. mit Bluetooth genauere Geschwindigkeitsangaben liefert als die Berechnung über die Android Location API bzw. die Messung über GPS.       

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas 
```

```python
# Gibt die Distanz in m zurück welche mit einem festen Zeitfenster t=timeframeT mit einer Geschwindigkeit in km/h zurückgelegt wurde
def distanceTraveledStaticTimeframe(kmh,timeframeT):
    return (kmh/3.6)*(timeframeT/1000.0)
# Die Funktion gibt die zurückgelegte Distanz in m während der fahrt zurück, wenn das Zeitfenster dynamisch berechnet wird 
def distanceTraveledDynamicTimeframe(kmh):
    return (kmh/3.6)*(1/(kmh/3.6))
# Berechnet zu der gegebenen Geschwindigkeit das dynamische Zeitfenster in ms
def dynamicTimeframeCalcT(kmh):
    return (1/(kmh/3.6))*1000
# Berechnet die Anzahl der Messwerte eines Zeitfensters t.
# timeframe: Zeitfenster in ms
# period: Periode in ms
def sensorSamples(timeframe, period):
    return timeframe/period
```

```python
# Diagramm m / km/h abhängig von der Dauer t des Zeitfensters
kmhArray = np.arange(5.0, 70.1, 0.01)
plt.figure(figsize=(15,8))
timeframe500ms, = plt.plot(kmhArray, distanceTraveledStaticTimeframe(kmhArray,500.0), 'k', label='Zeitfenster $t=500ms$')
timeframe80ms, = plt.plot(kmhArray, distanceTraveledStaticTimeframe(kmhArray,80.0), 'b', label='Zeitfenster $t=80ms$')
dynamicTimeframe, = plt.plot(kmhArray, distanceTraveledDynamicTimeframe(kmhArray), 'g', label='Dynamisches Zeitfenster $t=[9,432ms,720ms]$')
plt.grid()
plt.xlabel('Geschwindigkeit $v$ in $km/h$')
plt.ylabel('Weg $l$ in $m$')
plt.xlim(5,70)
plt.ylim(0,10)
plt.xticks(np.arange(5.0, 70.01, 5.0))
plt.yticks(np.arange(0.0, 10.01, 1.0))
plt.title('Zurückgelegte Distanz in Abhängigkeit zu der Dauer des Zeitfensters und der Geschwindigkeit')
plt.legend(handles=[timeframe500ms,timeframe80ms,dynamicTimeframe], loc=5)
plt.savefig('timeframeLengthDiagram')
plt.show()
```

```python
plt.figure(figsize=(15,8))
dynamicTimeframeLabel, = plt.plot(kmhArray, dynamicTimeframeCalcT(kmhArray), 'b', label='Dynamisches Zeitfenster $t=[9,432ms,720ms]$')
plt.grid()
plt.xlabel('Geschwindigkeit $v$ in $km/h$')
plt.ylabel('Zeitfenster $t$ in $ms$')
plt.xlim(5.0,70.0)
plt.ylim(0,1000)
plt.title('Dauer des dynamischen Zeitfensters gegenüber der Geschwindigkeit')
plt.legend(handles=[dynamicTimeframeLabel], loc=1)
plt.show()
```

```python
plt.figure(figsize=(15,8))
countSensorValues = sensorSamples(dynamicTimeframeCalcT(kmhArray),8)
timeframeArray500 = np.empty(kmhArray.shape[0])
timeframeArray500.fill(500)
timeframeArray80 = np.empty(kmhArray.shape[0])
timeframeArray80.fill(80)
timeframe500Label, = plt.plot(kmhArray, sensorSamples(timeframeArray500,8), 'k', label='Zeitfenster $t=500ms$')
timeframe80Label, = plt.plot(kmhArray, sensorSamples(timeframeArray80,8), 'b', label='Zeitfenster $t=80ms$')
dynamicTimeframeLabel, = plt.plot(kmhArray, countSensorValues, 'g', label='Dynamisches Zeitfenster $t=[9,432ms,720ms]$')
plt.grid()
plt.xlabel('Geschwindigkeit $v$ in $km/h$')
plt.ylabel('Anzahl der Sensormesswerte')
plt.xticks(np.arange(5.0, 70.01, 5.0))
plt.yticks(np.arange(0.0, 90.01, 5.0))
plt.xlim(5.0,70.0)
plt.ylim(0,90)
plt.title('Anzahl der Sensormesswerte mit einer Abtastrate von $f=125Hz$, $T=8ms$')
plt.legend(handles=[timeframe500Label,timeframe80Label,dynamicTimeframeLabel], loc=1)
plt.savefig('sensorSamples')
plt.show()
```
