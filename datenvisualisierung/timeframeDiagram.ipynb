# Zurückgelegter Weg in Abhängigkeit zur der Dauer des Zeitfensters

## Einleitung

Das Verhältnis zwischen zurückgelegter Strecke in $m$ während Messwerte mit den Smartphone Sensoren erfasst werden und Geschwindigkeit in $km/h$ wird in einem Diagramm dargestellt. Zu zeigen ist, dass die Länge $l$ der gefahrenen Strecke abhängig ist von der Dauer des Zeitfensters in $ms$. Die Annahme ist, dass bei der Wahl eines dynamischen Zeitfensters die zurückgelegte Strecke $l$ auf einen Meter normiert werden kann. Zudem wird ein Zeitfenster der Dauer $t=500ms$ und $t=80ms$ untersucht.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gibt die Distanz in m zurück welche mit einem festen Zeitfenster t=timeframeT mit einer Geschwindigkeit in km/h zurückgelegt wurde
def distanceTraveledStaticTimeframe(kmh,timeframeT):
    return (kmh/3.6)*(timeframeT/1000.0)
# Die Funktion gibt die zurückgelegte Distanz in m während der Fahr zurück, wenn das Zeitfenster dynamisch berechnet wird 
def distanceTraveledDynamicTimeframe(kmh):
    return (kmh/3.6)*(1/(kmh/3.6))
# Berechnet zu der gegebenen Geschwindigkeit das dynamische Zeitfenster in ms
def dynamicTimeframeCalcT(kmh):
    return (1/(kmh/3.6))*1000
```

```python
# Diagramm m / km/h abhängig von der Dauer t des Zeitfensters
kmhArray = np.arange(0.0, 200.0, 0.01)
plt.figure(figsize=(15,8))
timeframe500ms, = plt.plot(kmhArray, distanceTraveledStaticTimeframe(kmhArray,500.0), 'k', label='Zeitfenster t=500ms')
timeframe80ms, = plt.plot(kmhArray, distanceTraveledStaticTimeframe(kmhArray,80.0), 'b', label='Zeitfenster t=80ms')
dynamicTimeframe, = plt.plot(kmhArray, distanceTraveledDynamicTimeframe(kmhArray), 'g', label='Dynamisches Zeitfenster t=[0ms,720ms]')
plt.grid()
plt.xlabel('Geschwindigkeit in km/h')
plt.ylabel('Länge in m')
plt.xlim(5,70)
plt.ylim(0,10)
plt.title('Zurückgelegte Distanz in Abhängigkeit der Dauer des Zeitfensters und der Geschwindigkeit')
plt.legend(handles=[timeframe500ms,timeframe80ms,dynamicTimeframe], loc=5)
plt.show()
```

```python
plt.figure(figsize=(15,8))
dynamicTimeframeLabel, = plt.plot(kmhArray, dynamicTimeframeCalcT(kmhArray), 'b', label='Dynamisches Zeitfenster t=[0ms,720ms]')
plt.plot(kmhArray, e(kmhArray),'g', label='Dynamisches Zeitfenster t=[0ms,720ms]')
plt.grid()
plt.xlabel('Geschwindigkeit in km/h')
plt.ylabel('Zeitfenster ')
plt.xlim(0,200)
plt.ylim(0,100)
plt.title('Zurückgelegte Distanz in Abhängigkeit der Dauer des Zeitfensters und der Geschwindigkeit')
plt.legend(handles=[dynamicTimeframeLabel], loc=1)
plt.show()
```

```python
def e(x):
    return 0.1 * np.exp(-x) 
```
